## Data Transformation

Let's look at the unique values of categorical columns.

```python
categorical_vars = []

for col in data.select_dtypes(object):
    categorical_vars.append(col)
    print(f"{data[col]} : {data[col].unique()}")
    print("=======================================")
```

They all seem like ordinal categorical features. Let's specify that.

```python
for col in data.select_dtypes(object):
    data[col] = pd.Categorical(data[col], categories=data[col].unique(), ordered=True)

for col in categorical_vars:
    print(data[col].iloc[:4])
```

Good. Because they are ordinal categorical columns, we will encode them using label encoding.

```python
for col in categorical_vars:
    data[col] = data[col].astype('category').cat.codes
    
data.head()
```

The data looks good. We will now create the columns that we will use as a target column for classification.

```python
data['is_anomaly'] = data['attack'].map(lambda a: 0 if a == 0 else 1)

data.head()
```

```python
data.hist(figsize=(30, 30))
```

Good. Let's use our two machine learning algorithms before using feature selection.

## Modelling (Without Feature Selection)

First, we will create the method through which the models' metrics will be displayed.

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_scores(clf, X_train, X_test, y_train, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print('Training Results:')
        print('=================================================')
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.7f}%")
        print('=================================================')
        print(f"Classification Report: \n{clf_report}")
        print('=================================================')
        print(f"Confusion Matrix: \n{confusion_matrix(y_train, pred)}")
    else:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print('Test Results:')
        print('=================================================')
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.7f}%")
        print('=================================================')
        print(f"Classification Report: \n{clf_report}")
        print('=================================================')
        print(f"Confusion Matrix: \n{confusion_matrix(y_test, pred)}")
```

Next, we will split the dataset into a training dataset and a testing dataset.

```python
from sklearn.model_selection import train_test_split

X = data.drop('is_anomaly', axis=1)
y = data['is_anomaly']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Next, we will determine the right cross-validation score.

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

clf = DecisionTreeClassifier(random_state=42)
skf = StratifiedKFold(n_splits=10)

scores = cross_val_score(clf, X_train, y_train, cv=skf, scoring='f1')
print("Tree Score: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = GaussianNB()

scores = cross_val_score(clf, X_train, y_train, cv=skf, scoring='f1')
print("Bayes Score: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

Now, we will use two classifiers: decision trees, and naive bayes. Furthermore, we will employ automatic hyperparameter tuning.

```python
from sklearn.model_selection import GridSearchCV

tree_params = {
    "criterion":("gini", "entropy"), 
    "splitter":("best", "random"), 
    "max_depth":(list(range(1, 20))), 
    "min_samples_split":[2, 3, 4], 
    "min_samples_leaf":list(range(1, 20)),
}

bayes_params = {
    "var_smoothing": np.logspace(0, -15, 100)
}

dtc = DecisionTreeClassifier(random_state=42)

# 5-fold cross-validation
tree_cv = GridSearchCV(dtc, tree_params, scoring='recall', n_jobs=-1, verbose=1, cv=1)
tree_cv.fit(X_train, y_train)
tree_best_params = tree_cv.best_params_

dtc = DecisionTreeClassifier(**tree_best_params)
dtc.fit(X_train, y_train)
```

```python
nbc = GaussianNB()

# 5-fold cross-validation
bayes_cv = GridSearchCV(nbc, bayes_params, scoring='recall', n_jobs=-1, verbose=1, cv=1)
bayes_cv.fit(X_train, y_train)
bayes_best_params = bayes_cv.best_params_

nbc = GaussianNB(**bayes_best_params)
nbc.fit(X_train, y_train)
```

Let's see our scores.

```python
print("Decision Tree Scores:")
print()
print_scores(dtc, X_train, X_test, y_train, y_test, train=True)
print()
print_scores(dtc, X_train, X_test, y_train, y_test, train=False)
print()
print("==========================================")
print()
print("Naive Bayes Scores:")
print()
print_scores(nbc, X_train, X_test, y_train, y_test, train=True)
print()
print_scores(nbc, X_train, X_test, y_train, y_test, train=False)
```

## Data Reduction

Just like the previous time, we will use feature selection to select the relevant features.

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(f_classif, k=13)

selected_data = selector.fit_transform(X, y)

selected_data.shape
```

```python
mask = selector.get_support()
selected_features = []

for bool, feature in zip(mask, X.columns):
    if bool:
        selected_features.append(feature)
        
filtered_data = pd.DataFrame(selected_data, columns=selected_features)

filtered_data.head()
```

```python
filtered_data['is_anomaly'] = data['is_anomaly']

filtered_data.head()
```
